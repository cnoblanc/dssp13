```python
path=os.getcwd()
data_weather = pd.read_csv(os.path.join(path, 'submissions/dummies/external_data.csv'))
X_weather = data_weather[['Date', 'AirPort', 'Events','Mean TemperatureC','Dew PointC','Mean Humidity', \
                          'Mean Sea Level PressurehPa','Mean VisibilityKm','Mean Wind SpeedKm/h','CloudCover']].copy()
X_weather['Date_dt'] = pd.to_datetime(X_weather['Date'], format="%Y-%m-%d")
X_weather = X_weather.drop('Date', axis=1)
```

```python
#X_weather['Date_dt'] + datetime.timedelta(days=1)
```

```python
cal = calendar()
one_day=datetime.timedelta(days=1)
two_days=datetime.timedelta(days=2)
three_days=datetime.timedelta(days=3)
holidays = cal.holidays(start=X_weather['Date_dt'].min(), end=X_weather['Date_dt'].max()+ datetime.timedelta(days=10))
X_weather['isHoliday']=X_weather['Date_dt'].isin(holidays)
X_weather['isHoliday_one_day']=(X_weather['Date_dt']+one_day).isin(holidays)
X_weather['isHoliday_two_days']=(X_weather['Date_dt']+two_days).isin(holidays)
X_weather['isHoliday_three_days']=(X_weather['Date_dt']+three_days).isin(holidays)
X_weather['isHoliday_in_next_days']=(X_weather['isHoliday'] | X_weather['isHoliday_one_day'] \
                                   | X_weather['isHoliday_two_days']| X_weather['isHoliday_three_days']).astype(int)

X_weather
```
