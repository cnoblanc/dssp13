```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
```

## Example 1

```python
mu1 = [2, 2] # Position des moyennes (les centres)
mu2 = [-2, -3]
sigma1 = [[1, 0], [0, 1]] # Co-variance : matrice identité
sigma2 = [[1, 0], [0, 1]]

# Itère sur 3 variables en donnant chaque possibilité
for mu, sigma, color in zip((mu1, mu2), (sigma1, sigma2), ('r', 'b')): 
    X = np.random.multivariate_normal(mu, sigma, size=200) # Génère des X en gaussienne
    plt.scatter(X[:, 0], X[:, 1], color=color)
    plt.scatter(mu[0], mu[1], color='k', s=300., alpha=0.8, marker='h')

plt.axis('equal');

# On a envie de prendre la médiatrice pour séparer les 2 populations
```

## Example 2

```python
mu1 = [2, 2]
mu2 = [-6, -2]
sigma1 = [[7, 0], [0, 0.7]] # Variance de 7 en X, mais peu de variance sur Y
sigma2 = [[7, 0], [0, 0.7]] 

for mu, sigma, color in zip((mu1, mu2), (sigma1, sigma2), ('r', 'b')):
    X = np.random.multivariate_normal(mu, sigma, size=200)
    plt.scatter(X[:, 0], X[:, 1], color=color)
    plt.scatter(mu[0], mu[1], color='k', s=300., alpha=0.8, marker='h')

plt.axis('equal');

# On a envie de faire une rotation de la médiatrice pour séparer les 2 populations
```

## Example 3

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target

# Take only 2 features
X = X[:, :2]

# And add a column of 1 for the intercept
X = np.concatenate((X, np.ones((X.shape[0], 1))), axis=1)

# Make it binary
X = X[y < 2]
y = y[y < 2]

y[y == 0] = -1

plt.scatter(X[y > 0, 0], X[y > 0, 1], color='r', label=iris.target_names[0])
plt.scatter(X[y < 0, 0], X[y < 0, 1], color='g', label=iris.target_names[1])
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend();
```
